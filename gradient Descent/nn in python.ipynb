```python
coef1 =1.136335 
coef2= 0.7236735
bias =1.6098249
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def prediction(age , Salary):
    sum = coef1*age + coef2*Salary +bias
    return sigmoid(sum)

prediction(2,400)
```

```python
epsilon= 1e-15
```

```python
def Binary_Cross_Entropy(y_true,y_pred):
    y_pred_new= [min(i,1-epsilon) for i in y_pred]
    y_pred_new=[max(i,epsilon) for i in y_pred_new]
    y_pred_new= np.array(y_pred_new)
    return -np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))
```

```python
def gradient_descent(age,Salary, epoch, y_true):
    w1=w2=0
    bias=0
    rate = 0.1 #alpha
    n = len(age)
    for i in range(epoch):
        weighted_sum = w1*age + w2*Salary+bias
        y_pred = sigmoid(weighted_sum)
        
        loss = Binary_Cross_Entropy(y_true, y_pred)
        #derivatives
        w1d = (1/n)* np.dot(np.transpose(age) ,(y_pred -y_true))
        w2d = (1/n)* np.dot(np.transpose(Salary) ,(y_pred -y_true))
        
        biasd= np.mean(y_pred - y_true)
        w1= w1 - rate *w1d
        w2 =w2 -rate*w2d
        bias = bias -rate*biasd
        print(f"Epoch{i}: w1={w1} ,w2 ={w2},  bias ={bias}, loss = {loss}")
    return w1,w2,bias
```
